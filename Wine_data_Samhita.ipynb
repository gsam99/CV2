```python
import pandas as pd
sensory_data = pd.read_excel("/content/drive/MyDrive/Wine color data.xlsx",sheet_name="Sensory evaluation - human eyes",skiprows=lambda x: x % 8 == 2 or x % 8 == 3)

instrumental_data=pd.read_excel("/content/drive/MyDrive/Wine color data.xlsx",sheet_name="Instrument analysis - spec")
```

```python
sensory_data.head()
```

```python
instrumental_data.head()
```

```python
sensory_data["Replicate"]=sensory_data["Replicate"].str.replace("R","",regex=False)
```

```python
sensory_data
```

```python
print(sensory_data.describe())
```

```python
print(sensory_data.info())
```

```python
sensory_data["Replicate"]=sensory_data["Replicate"].astype('int')
```

```python
import pandas as pd
merged_data=pd.merge(sensory_data,instrumental_data,on=['Treatment','Replicate'],how='inner')
```

```python
merged_data.head()
```

```python
merged_data.info()
```

```python
from sklearn.model_selection import train_test_split
X=merged_data[["Replicate","Panelist no. ","Hue","Color intensity"]]
Y=merged_data[["Panelist - Hue","Panelist - Color intensity"]]
print(X.shape)
print(Y.shape)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=123)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
X_train
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(units=64,activation='relu',input_dim=X_train.shape[1]))
model.add(Dense(units=32,activation='relu'))
model.add(Dense(units=16,activation='relu'))
model.add(Dense(units=8,activation='relu'))
model.add(Dense(units=2))
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
model.fit(X_train,Y_train,epochs=50,batch_size=4)
```

```python
loss,mae=model.evaluate(X_test,Y_test)
print(f'MSE:{loss}')
print(f'MAE:{mae}')
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
loss, mae = model.evaluate(X_test, Y_test, verbose=0)
print(f"MSE (overall): {loss}")
print(f"MAE (overall): {mae}")
predictions = model.predict(X_test)
y_true_hue = Y_test.iloc[:, 0]
y_true_intensity = Y_test.iloc[:, 1]
y_pred_hue = predictions[:, 0]
y_pred_intensity = predictions[:, 1]
rmse_hue = np.sqrt(mean_squared_error(y_true_hue, y_pred_hue))
rmse_intensity = np.sqrt(mean_squared_error(y_true_intensity, y_pred_intensity))
r2_hue = r2_score(y_true_hue, y_pred_hue)
r2_intensity = r2_score(y_true_intensity, y_pred_intensity)
print(f"Hue RMSE: {rmse_hue:.4f}, RÂ²: {r2_hue:.4f}")
print(f"Color Intensity RMSE: {rmse_intensity:.4f}, RÂ²: {r2_intensity:.4f}")
```
